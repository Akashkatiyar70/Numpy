```python
# logistic distribution - describe growth it is basically important in regression and neural network.

# param  ==> loc(mean-0), scale(standard deviation,1),size

# draw 2*2 samples of logistic with mean at 1 and stdev 2.0

from numpy import random
ak =random.logistic(loc=1,scale=2,size=(2,3))
print(ak)
```

```python
# visulaization of logistics dist
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.logistic(size=1000),hist=False)
plt.show()
```
