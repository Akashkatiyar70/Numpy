```python
# Normal Distribution(Gaussian)- very important.
# random.normal() method - loc, scale, size

# we are generating a random normal distribution of size 2*3
from numpy import random
ak = random.normal(size=(2,3))
print(ak)
```

```python
# here ew will generating a random normal dist of size 2*3 with mean at 1 and standard deviation of 2:

from numpy import random
ak = random.normal(loc=1, scale=2,size=(2,3))
print(ak)
```

```python
# visualization of normal distribution
from numpy import random 
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(size=100),hist=False)
plt.show()

# The curve of a normal dist is called as bell curve
```
